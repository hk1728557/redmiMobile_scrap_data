```python
items = [] #to create list
base_url = "https://www.flipkart.com/mobiles/mi~brand/pr?sid=tyy%2C4io&otracker=nmenu_sub_Electronics_0_Mi&page={}"

for page_no  in tqdm(range(1,total_pages+1,1)): #set the start page , end page and incrment
    #print(page_no)
    page_url = base_url.format(page_no) #to set the page url formate put page number in base url 
    #print(page_url)
    res = requests.get(page_url)  #send the request to the url and check respnse code
    sel = Selector(res)           # using selector to scrap the page
    
    # set all scrap data path using xpath
    prod_url = sel.xpath('//*/a[@class="_1fQZEK"]/@href').extract()
    prod_title = sel.xpath('//*//div[contains(@class,"_4rR01T")]//text()').extract()
    prod_price = sel.xpath('//*//div[contains(@class,"_30jeq3 _1_WHN1")]//text()').extract()
    prod_rating =sel.xpath('//*/span[@class="_1lRcqv"]/div/text()').extract()
    attri1 = sel.xpath('///*/ul[@class="_1xgFaf"]/li[1]/text()').extract()
    attri2 = sel.xpath('//*/ul[@class="_1xgFaf"]/li[2]/text()').extract()
    attri3 = sel.xpath('///*/ul[@class="_1xgFaf"]/li[3]/text()').extract()
    attri4 = sel.xpath('///*/ul[@class="_1xgFaf"]/li[4]/text()').extract()
    attri5 = sel.xpath('///*/ul[@class="_1xgFaf"]/li[5]/text()').extract()
    
    #now uing for loop zip all the products values
    for prod_url,prod_title,prod_price,prod_rating,attri1,attri2,attri3,attri4,attri5 in zip(prod_url,prod_title,prod_price,prod_rating,attri1,attri2,attri3,attri4,attri5):
        
        item = {}        #create dict and put all the valu in dict.
        item['list_view_page_url'] = page_url
        item['prod_url'] = prod_url 
        item['prod_title'] = prod_title
        item[' prod_price'] = prod_price
        item['prod_rating'] = prod_rating
        item['attri1'] = attri1
        item['attri2'] = attri2
        item['attri3'] = attri3
        item['attri4'] = attri4
        item['attri5'] = attri5
        
        items.append(item) #now append all the items list  value to the item dictionary
    
    
```

```python
df = pd.DataFrame.from_records(items) #create data frame using pandas 
df
```
